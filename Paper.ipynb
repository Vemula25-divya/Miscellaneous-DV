```python
introduction = """
1. INTRODUCTION

Pseudomonas aeruginosa is a ubiquitous Gram-negative opportunistic
pathogen of significant clinical concern, responsible for severe
nosocomial infections including ventilator-associated pneumonia,
urinary tract infections, bloodstream infections, and chronic
pulmonary infections in cystic fibrosis patients [1]. The World
Health Organization (WHO) has classified carbapenem-resistant
P. aeruginosa as a critical priority pathogen for which new
antibiotics are urgently needed [2]. The intrinsic resistance
mechanisms of P. aeruginosa, including low outer membrane
permeability, constitutive expression of efflux pumps, and
remarkable capacity for acquiring resistance determinants through
horizontal gene transfer, render it one of the most difficult
pathogens to treat clinically [3]. The limited efficacy of
existing antibiotics and the alarming rise of multidrug-resistant
(MDR) and extensively drug-resistant (XDR) strains have created
a critical need for novel therapeutic strategies targeting
previously unexploited pathways [4].

Lipid A, the hydrophobic anchor of lipopolysaccharide (LPS), is
an essential structural component of the outer membrane of
Gram-negative bacteria and plays a crucial role in bacterial
viability and outer membrane integrity [5]. The biosynthesis of
lipid A proceeds through a conserved nine-step pathway, of which
the deacetylation of UDP-3-O-acyl-N-acetylglucosamine — catalyzed
by LpxC (UDP-3-O-acyl-N-acetylglucosamine deacetylase) — represents
the first committed and rate-limiting step [6]. LpxC is a
zinc-dependent metalloamidase with no known mammalian homolog,
making it an attractive target for the development of
pathogen-selective antibacterial agents with minimal host
toxicity [7]. Genetic studies have confirmed the essentiality
of LpxC for bacterial growth and survival in Gram-negative
organisms, further validating it as a high-priority therapeutic
target [8]. Several LpxC inhibitors, notably hydroxamic
acid-based compounds such as CHIR-090 and LpxC-4, have
demonstrated potent antibacterial activity against P. aeruginosa
in preclinical studies, confirming the druggability of this
target [9,10].

Despite the promise of LpxC as a drug target, the development
of clinically viable LpxC inhibitors has been challenging due
to the structural diversity required to achieve selectivity,
potency, and favorable pharmacokinetic properties simultaneously.
Traditional drug discovery approaches are time-consuming,
resource-intensive, and associated with high attrition rates.
Computational approaches, particularly machine learning-based
quantitative structure-activity relationship (QSAR) modeling
and structure-based virtual screening, have emerged as powerful
tools to accelerate the early stages of drug discovery by
enabling rapid prioritization of large compound libraries [11,12].

Machine learning algorithms, including Random Forest and gradient
boosting methods, have demonstrated remarkable success in
predicting the biological activity of small molecules from
molecular descriptors and fingerprints [13,14]. These methods
can capture complex nonlinear structure-activity relationships
that are difficult to model using classical statistical
approaches. Furthermore, the integration of interpretability
tools such as SHapley Additive exPlanations (SHAP) enables
mechanistic understanding of model predictions, providing
valuable insights into the molecular features driving biological
activity [15]. Combined with pharmacophore-based filtering and
structure-based molecular docking, ML-based virtual screening
pipelines offer a cost-effective and efficient approach for
identifying novel hit compounds from large and diverse chemical
spaces [16].

Several computational studies have investigated LpxC inhibitors
using QSAR and molecular docking approaches; however, most
have been limited in scope, focusing on small compound sets or
single chemical libraries [17,18]. A comprehensive ML-based
virtual screening campaign integrating multiple diverse chemical
libraries — including natural product databases, macrocyclic
collections, and drug-like compound libraries — with mechanistic
pharmacophore filtering has not been reported for LpxC to date.

In the present study, we report the development and validation
of a robust ML-based cheminformatics model for the prediction
of LpxC inhibitory activity against P. aeruginosa, trained on
492 experimentally validated compounds from the ChEMBL database.
The model, built using Random Forest and XGBoost algorithms on
a combined feature set of RDKit molecular descriptors and Morgan
fingerprints, was rigorously validated through five-fold cross
validation and external validation. The validated Random Forest
model was deployed for large-scale virtual screening of over
1.3 million compounds across six diverse chemical libraries,
preceded by zinc binding group pharmacophore filtering to
ensure mechanistic relevance. High-confidence hits were further
evaluated through ADMET profiling and Glide XP molecular
docking. The integrated computational pipeline identified
several structurally novel LpxC inhibitor candidates, including
natural product-derived scaffolds not previously reported as
LpxC inhibitors, providing a strong foundation for future
experimental validation and medicinal chemistry optimization.
"""

with open(project_path + r'\paper\introduction.txt',
          'w', encoding='utf-8') as f:
    f.write(introduction)

print("Introduction saved ✅")
print(f"Word count: {len(introduction.split())} words")
print("\n" + introduction)
```

```python
methods = """
2. MATERIALS AND METHODS

2.1 Dataset Collection and Preparation

Bioactivity data for LpxC (UDP-3-O-acyl-N-acetylglucosamine
deacetylase) from Pseudomonas aeruginosa (strain ATCC 15692 /
PAO1) was retrieved from the ChEMBL database (version 33) using
the ChEMBL web resource client in Python [19]. The target was
identified using ChEMBL target ID CHEMBL3855. Compounds with
IC50 values measured against this specific target were collected,
yielding an initial dataset of 588 records. The following
preprocessing steps were applied: (i) only compounds with IC50
values reported in nanomolar (nM) units were retained; (ii)
records with non-exact measurements (standard relation != '=')
were excluded; (iii) duplicate entries based on ChEMBL molecule
ID were removed; and (iv) compounds with missing SMILES or IC50
values were discarded. After preprocessing, 492 compounds were
retained for model development.

IC50 values were converted to the negative logarithmic scale
(pIC50) using the formula:

    pIC50 = -log10(IC50 x 10^-9)

where IC50 is expressed in nanomolar units. Compounds were
labeled as active (pIC50 >= 6.0, corresponding to IC50 <= 1000
nM) or inactive (pIC50 < 6.0) for classification modeling,
yielding 398 active and 94 inactive compounds. This threshold
is consistent with established QSAR literature for LpxC
inhibitors [20].

2.2 Molecular Descriptor Calculation and Feature Engineering

Molecular structures were standardized and validated using
RDKit (version 2025.09.5) [21]. SMILES strings were converted
to RDKit molecule objects, and invalid SMILES were identified
and removed. A total of 217 molecular descriptors were
calculated for each compound using the RDKit MolecularDescriptor-
Calculator, encompassing constitutional, topological,
physicochemical, and electronic descriptor categories.
Additionally, 2048-bit Morgan fingerprints (circular
fingerprints) with radius 2 were generated using the
GetMorganFingerprintAsBitVect function, encoding the presence
or absence of circular substructural features up to two bonds
from each atom.

To reduce redundancy in the descriptor set, Pearson correlation
analysis was performed and descriptors exhibiting pairwise
correlation coefficients exceeding 0.95 were identified. From
each highly correlated pair, one descriptor was removed,
yielding a final set of 179 non-redundant descriptors. The
combined feature matrix consisting of 179 RDKit descriptors
and 2048 Morgan fingerprint bits (2227 total features) was
used for model training and virtual screening.

2.3 Machine Learning Model Development

Two ensemble machine learning algorithms were evaluated for
LpxC inhibitor classification: Random Forest (RF) and
Extreme Gradient Boosting (XGBoost). The dataset was split
into training (80%, n=393) and test (20%, n=99) sets using
stratified random splitting to preserve the active/inactive
ratio. Class imbalance (398 active, 94 inactive) was addressed
using the class_weight='balanced' parameter in Random Forest
and scale_pos_weight in XGBoost.

The Random Forest classifier was trained with 100 decision
trees (n_estimators=100), while XGBoost was configured with
100 boosting rounds and binary logistic objective. Both models
were implemented using scikit-learn (version 1.7.2) and
XGBoost (version 2.1.3) libraries in Python.

2.4 Model Validation

Model performance was evaluated using multiple metrics
including accuracy, ROC-AUC, precision, recall, and F1-score.
Internal validation was performed using stratified 5-fold
cross-validation to assess model robustness and generalizability.
External validation was conducted on an independent set of
42 P. aeruginosa LpxC inhibitors not included in the training
dataset. For regression modeling, a Random Forest Regressor
was trained to predict continuous pIC50 values, evaluated
using R-squared (R²), root mean squared error (RMSE), and
mean absolute error (MAE).

2.5 Feature Importance and SHAP Analysis

Feature importance was assessed using two complementary
approaches: (i) built-in feature importance scores from both
Random Forest and XGBoost models, which rank descriptors by
their contribution to prediction accuracy; and (ii) SHapley
Additive exPlanations (SHAP) analysis using the TreeExplainer
module of the SHAP library (version 0.49.1) [22]. SHAP values
provide a unified framework for interpreting model predictions
by quantifying the contribution of each feature to individual
predictions. Consensus descriptors identified by all three
methods (RF importance, XGBoost importance, and SHAP) were
considered the most reliable molecular features governing
LpxC inhibition.

2.6 Zinc Binding Group Pharmacophore Filtering

Prior to virtual screening, a mechanistic pharmacophore filter
was applied to all compound libraries to retain only compounds
containing at least one zinc binding group (ZBG) capable of
coordinating the catalytic zinc ion in the LpxC active site.
Nine ZBG patterns were defined using SMARTS notation and
implemented using RDKit substructure matching:
(i) hydroxamic acid [CX3](=O)[NX3][OH],
(ii) carboxylic acid [CX3](=O)[OH],
(iii) thiol [SX2H],
(iv) phosphonate [PX4](=O)([OH])[OH],
(v) sulfonamide [SX4](=O)(=O)[NX3],
(vi) catechol c1ccc(O)c(O)c1,
(vii) N-hydroxy [NX3][OH],
(viii) hydroxypyridine n1ccccc1O, and
(ix) beta-lactam [NX3]1[CX4][CX3](=O)1.
Compounds were prioritized based on ZBG type, with hydroxamic
acid-containing compounds receiving highest priority given
their established role as LpxC inhibitor pharmacophores.

2.7 Virtual Screening Pipeline

The validated Random Forest model was deployed for large-scale
virtual screening across six diverse chemical libraries:
(i) ChEMBL subsets including natural products, antibacterials,
antiviral compounds (ATC J05), anticancer agents (ATC L),
cardiovascular drugs (ATC C), antiparasitic compounds (ATC P),
terpenoids, alkaloids, flavonoids, glycosides, polyketides,
and macrocyclic compounds (total ~400,000 compounds);
(ii) COCONUT natural product database (721,006 compounds) [23];
(iii) OTAVA Drug-Like Green Collection (160,812 compounds);
(iv) Selleckchem Natural Product Library (3,671 compounds);
(v) MACROx macrocyclic compound collection (4,063 compounds);
and (vi) NATx natural product-inspired library (34,911 compounds).

For each library, SMILES strings were processed through RDKit
for descriptor and fingerprint calculation. Compounds were
first filtered using the ZBG pharmacophore filter, followed
by ML-based activity prediction using the trained Random Forest
classifier. Compounds with predicted active probability >= 0.85
were designated as high-confidence hits and prioritized for
further evaluation.

2.8 ADMET Profiling

Drug-likeness and ADMET (absorption, distribution, metabolism,
excretion, and toxicity) properties were predicted for
high-confidence hits using RDKit-calculated physicochemical
descriptors. Lipinski's Rule of Five (Ro5) compliance was
assessed based on molecular weight (<= 500 Da), lipophilicity
(LogP <= 5), hydrogen bond donors (<= 5), and hydrogen bond
acceptors (<= 10) [24]. Oral bioavailability was further
evaluated using Veber's rules (rotatable bonds <= 10 and
topological polar surface area <= 140 A²) [25]. Drug-likeness
was quantified using the Quantitative Estimate of Drug-likeness
(QED) score implemented in RDKit [26]. A composite drug score
was calculated by integrating Lipinski compliance, Veber
compliance, and QED score with weighted contributions.

2.9 Molecular Docking

High-confidence hits passing ADMET filters were subjected to
molecular docking using Glide XP (Extra Precision) [27]
implemented in Schrodinger Maestro (version 2024). The
crystal structure of P. aeruginosa LpxC was retrieved from
the Protein Data Bank and prepared using the Protein
Preparation Wizard, including addition of hydrogens, assignment
of bond orders, and restrained minimization using the OPLS4
force field. Ligands were prepared using LigPrep with Epik
ionization at pH 7.0 +/- 2.0. The receptor grid was generated
centered on the LpxC active site. Glide XP docking was
performed with flexible ligand sampling, and up to five poses
per ligand were retained. Docking poses were evaluated based
on Glide XP GScore, and binding free energies were estimated
using the MM-GBSA method.

2.10 Composite Scoring and Hit Prioritization

Final hit prioritization was performed using a composite
scoring approach integrating three orthogonal computational
filters: (i) ML predicted active probability (weight: 0.30),
(ii) normalized Glide XP docking score (weight: 0.35), and
(iii) normalized MM-GBSA binding free energy (weight: 0.35).
All scores were normalized to a 0-1 scale using MinMaxScaler
prior to combination. Structural novelty relative to known
LpxC inhibitors was assessed using Tanimoto similarity
calculated from 2048-bit Morgan fingerprints, with compounds
exhibiting Tanimoto similarity < 0.40 classified as novel
scaffolds. Final priority compounds were selected based on
composite score, structural novelty, ZBG type, and ADMET
compliance.

2.11 Software and Computational Environment

All cheminformatics analyses were performed in Python 3.10
using the following libraries: RDKit 2025.09.5, scikit-learn
1.7.2, XGBoost 2.1.3, SHAP 0.49.1, pandas 2.3.3, numpy
1.23.5, matplotlib 3.10.8, and seaborn 0.13.2. ChEMBL data
was accessed using the chembl-webresource-client library.
Molecular docking was performed using Schrodinger Glide XP.
All analyses were conducted on a Windows 10 workstation
with Intel Core i7 processor, 16 GB RAM, and NVIDIA GPU.
"""

with open(project_path + r'\paper\methods.txt',
          'w', encoding='utf-8') as f:
    f.write(methods)

print("Methods section saved ✅")
print(f"Word count: {len(methods.split())} words")
print("\n" + methods)
```

```python
results = """
3. RESULTS AND DISCUSSION

3.1 Dataset Characterization

A total of 492 compounds with experimentally validated IC50
values against P. aeruginosa LpxC (CHEMBL3855) were retrieved
from the ChEMBL database after preprocessing. The dataset
comprised 398 active compounds (pIC50 >= 6.0, 81.0%) and 94
inactive compounds (pIC50 < 6.0, 19.0%), reflecting the
enrichment of potent inhibitors in the published literature.
The pIC50 values ranged from 2.56 to 10.70, with a mean of
7.71 and standard deviation of 1.62, indicating a broad potency
range across the dataset. The distribution exhibited a bimodal
pattern with distinct active and inactive populations separated
at the pIC50 = 6.0 threshold (Figure 1), confirming the
suitability of this cutoff for binary classification.

The class imbalance (81% active vs 19% inactive) is inherent
to LpxC inhibitor datasets in the literature, as published
studies predominantly report potent inhibitors. This imbalance
was addressed during model training using class weighting
strategies to prevent model bias toward the majority class.

3.2 Molecular Descriptor Analysis and Feature Selection

A total of 217 molecular descriptors were calculated for all
492 compounds using RDKit. Pearson correlation analysis
revealed 38 descriptor pairs with correlation coefficients
exceeding 0.95, indicating significant redundancy predominantly
among size-related descriptors including molecular weight,
heavy atom count, and topological surface area variants.
Following removal of one descriptor from each highly correlated
pair, 179 non-redundant descriptors were retained. Combined
with 2048-bit Morgan fingerprints, the final feature matrix
comprised 2227 features per compound.

Comparative analysis of key physicochemical descriptors
between active and inactive compounds revealed statistically
meaningful differences in topological polar surface area
(TPSA; active median ~105 vs inactive ~80), number of
hydrogen bond donors (active median ~2 vs inactive ~1), and
number of rotatable bonds (active median ~7 vs inactive ~5),
suggesting that active LpxC inhibitors tend to be more polar
and flexible than inactive compounds (Figure 2). These
observations are consistent with the polar, zinc-coordinating
nature of the LpxC active site.

3.3 Machine Learning Model Performance

Both Random Forest and XGBoost classifiers demonstrated strong
predictive performance on the held-out test set (Table 1).
The Random Forest model achieved an accuracy of 96.97% and
ROC-AUC of 0.9806 on the test set, correctly predicting 96
out of 99 test compounds. XGBoost achieved an accuracy of
92.93% and ROC-AUC of 0.9842. The Random Forest model
demonstrated perfect recall for active compounds (1.00),
meaning no active compound was misclassified as inactive,
which is particularly important in virtual screening
applications where missing active compounds represents a
significant cost.

Table 1. Performance metrics of Random Forest and XGBoost
classifiers on the test set and five-fold cross-validation.

Metric               | Random Forest | XGBoost
---------------------|---------------|--------
Accuracy (test)      | 96.97%        | 92.93%
ROC-AUC (test)       | 0.9806        | 0.9842
Accuracy (5-fold CV) | 95.94±0.63%   | 94.52±1.36%
ROC-AUC (5-fold CV)  | 0.9834±0.0088 | 0.9856±0.0086
F1-score (Active)    | 0.98          | 0.96
F1-score (Inactive)  | 0.91          | 0.83
Active Recall        | 1.00          | 0.94

Five-fold stratified cross-validation confirmed the
robustness of both models, with the Random Forest model
exhibiting lower variance (±0.63%) compared to XGBoost
(±1.36%), indicating greater stability across different
data partitions. The ROC curves for both models demonstrated
excellent discrimination ability, with both curves closely
approaching the top-left corner of the ROC space (Figure 3).
Based on superior accuracy, active recall, and cross-validation
stability, the Random Forest model was selected for subsequent
virtual screening applications.

Regression modeling for continuous pIC50 prediction yielded
an R² of 0.7731, RMSE of 0.8143, and MAE of 0.5849 on the
internal test set, indicating that the model explains
approximately 77% of the variance in LpxC inhibitory potency.
The predicted vs actual pIC50 plot demonstrated good agreement
in the high-activity region (pIC50 8-10), with greater
deviation observed for lower-potency compounds (pIC50 < 6),
consistent with the lower density of inactive compounds in
the training set (Figure 4).

3.4 External Validation

External validation on 42 independent P. aeruginosa LpxC
inhibitors confirmed the generalizability of the Random Forest
model, achieving a classification accuracy of 90.48% and
ROC-AUC of 0.8293. Of the 42 external compounds, 38 were
correctly classified (38/42, 90.5%). The four misclassified
compounds comprised three active compounds predicted as
inactive (STR 52, STR 53, STR 72) and one borderline inactive
compound predicted as active (STR 65, pIC50 = 5.7). Analysis
of the misclassified active compounds revealed they belonged
to the STR series, a structurally distinct scaffold series
not well represented in the training data, highlighting the
known limitation of ML models with respect to applicability
domain. The regression model showed reduced performance on
the external set (R² = -3.36), attributable to the narrow
potency range of the external compounds (pIC50 5.7-10.7),
which represents an applicability domain limitation rather
than model failure.

3.5 Feature Importance and SHAP Analysis

Feature importance analysis using three independent methods
— Random Forest importance, XGBoost importance, and SHAP
analysis — converged on a consistent set of molecular
descriptors governing LpxC inhibition. BCUT2D_LOGPLOW, a
BCUT descriptor encoding lipophilicity distribution across
the molecular connectivity matrix, emerged as the most
important descriptor in all three analyses, with a Random
Forest importance of 0.105, XGBoost importance of 0.441,
and mean absolute SHAP value of 0.058 (Figure 5). This
finding demonstrates remarkable consistency across
fundamentally different algorithmic approaches, providing
strong evidence for the mechanistic relevance of
lipophilicity distribution in LpxC inhibitor design.

Seven consensus descriptors were identified across all
three feature importance methods: BCUT2D_LOGPLOW,
MinEStateIndex, MaxPartialCharge, Ipc, SMR_VSA5,
EState_VSA1, and SlogP_VSA8. These descriptors collectively
capture three molecular property themes relevant to
LpxC inhibition: (i) lipophilicity distribution
(BCUT2D_LOGPLOW, SlogP_VSA8), (ii) electronic properties
and partial charge distribution (MinEStateIndex,
MaxPartialCharge, EState_VSA1), and (iii) molecular
complexity and shape (Ipc, SMR_VSA5). The predominance
of electronic and lipophilicity-distribution descriptors
is mechanistically consistent with the zinc-dependent
catalytic mechanism of LpxC, where inhibitor binding
involves coordination with the active site zinc ion
through electronegative donor atoms.

SHAP beeswarm analysis further revealed the directionality
of descriptor contributions: high BCUT2D_LOGPLOW values
strongly push predictions toward active, while high values
of MinEStateIndex and BCUT2D_CHGHI push predictions toward
inactive (Figure 6). The SHAP dependence plot for
BCUT2D_LOGPLOW revealed a threshold effect at approximately
-2.5, below which compounds are consistently predicted
as active regardless of other features. Notably, the
presence of sulfonamide groups (fr_sulfone) acted as a
rescue feature, maintaining active predictions even at
suboptimal BCUT2D_LOGPLOW values, suggesting sulfonamide
as a key pharmacophoric element for LpxC inhibition.

3.6 Zinc Binding Group Pharmacophore Analysis

Application of the ZBG pharmacophore filter across all
screened libraries revealed the prevalence of different
zinc-coordinating functional groups. Across the 1.3 million
screened compounds, carboxylic acids and catechols were
the most abundant ZBGs, while hydroxamic acids — the
gold-standard ZBG for metalloenzyme inhibitors — were
present in smaller but highly enriched subsets. Hydroxamic
acid-containing compounds demonstrated consistently high
ML-predicted activity probabilities, validating the
mechanistic relevance of this pharmacophore filter for
LpxC inhibitor identification. In the COCONUT dataset,
400 hydroxamic acid compounds were identified, of which
278 (69.5%) were predicted active by the Random Forest
model, compared to 53.4% and 89.9% hit rates for
N-hydroxy/thiol and sulfonamide compounds respectively,
supporting the hypothesis that hydroxamic acid-bearing
natural products represent a privileged scaffold for
LpxC inhibition.

3.7 Virtual Screening Results

Large-scale virtual screening of over 1.3 million compounds
across six chemical libraries, preceded by ZBG pharmacophore
filtering, yielded a total of high-confidence hits
(predicted probability >= 0.85) from each library.
The screening results are summarized in Table 2.

Table 2. Summary of virtual screening results across
all chemical libraries.

Library              | Screened  | ZBG pass | High conf hits
---------------------|-----------|----------|---------------
ChEMBL subsets       | ~400,000  | N/A      | 132
COCONUT              | 721,006   | 120,627  | 278 (HA only)
OTAVA Drug-Like      | 160,812   | 38,486   | 108
Selleckchem NP       | 3,671     | 1,326    | 37
MACROx               | 4,063     | 1,517    | 32
NATx                 | 34,911    | 6,778    | 141
Total                |>1,300,000 |          | 728+

Cross-library analysis revealed consistent enrichment of
specific scaffold classes across multiple independent
libraries, providing strong validation of the model's
predictive utility. Cephalosporin antibiotics
(ceftazidime, moxalactam, cefiderocol, cefazedone)
appeared as high-confidence hits across ChEMBL, Selleckchem,
and FDA-approved drug libraries, consistent with their
known beta-lactam pharmacophore and carboxylic acid ZBG.
Protease inhibitors (ritonavir, boceprevir, narlaprevir,
telaprevir) demonstrated unexpected but structurally
consistent predictions, suggesting scaffold overlap with
known LpxC inhibitors. Among natural product libraries,
the Casear- series diterpenoids from marine sources and
isochlorogenic acid derivatives from plant sources emerged
as structurally novel scaffolds with no prior reported
LpxC activity.

Structural novelty analysis using Tanimoto similarity
against the 398 training set actives revealed that 174
out of 178 representative hits (97.8%) exhibited maximum
Tanimoto similarity below 0.40, confirming that the
screening identified genuinely novel chemical space beyond
the training data. The mean maximum Tanimoto similarity
of 0.202 across all hits indicates that the model
successfully extrapolated to structurally distant
compound classes.

3.8 ADMET Analysis

ADMET profiling of 46 representative high-confidence hits
from all libraries revealed that compounds from the
docking hit set demonstrated the most favorable drug-like
profiles, with six compounds passing all filters including
Lipinski Ro5, Veber rules, QED > 0.40, and composite
drug score > 0.80. STK989244 emerged as the top drug-like
candidate with molecular weight of 319.4 Da, LogP of 3.88,
TPSA of 52.49, QED of 0.646, and high GI absorption
prediction. The majority of natural product hits showed
poor Veber compliance due to high TPSA and rotatable bond
counts, consistent with the known pharmacokinetic challenges
of natural product scaffolds. FDA-approved antibiotics
(ceftazidime, dalfopristin) showed poor ADMET scores,
expected given their intravenous administration routes
rather than oral delivery.

3.9 Molecular Docking Results

High-confidence hits from COCONUT, OTAVA, Selleckchem,
MACROx, and NATx libraries were subjected to Glide XP
molecular docking against the P. aeruginosa LpxC crystal
structure. Among the docked compounds, PLANTAGINOSIDE A
demonstrated the strongest binding with a Glide XP GScore
of -11.607 kcal/mol, followed by ISOQUERCETIN (-10.595)
and PLANTAINOSIDE D (-10.593). Isochlorogenic acid
derivatives and caffeoylquinic acids from the Selleckchem
Natural Product Library showed consistently strong docking
scores (GScore range -9.0 to -11.0), with dual carboxylic
acid-catechol ZBGs potentially enabling bidentate zinc
coordination. Hydroxamic acid-bearing COCONUT compounds
demonstrated favorable binding geometries with predicted
zinc coordination distances consistent with metalloenzyme
inhibition. The composite scoring approach integrating
ML probability, docking score, and MM-GBSA binding free
energy identified STK651137 as the overall top-ranked
compound with the most favorable MM-GBSA binding energy
of -52.5 kcal/mol combined with ML probability of 0.882.

3.10 Identification of Priority Compounds

Integration of all computational filters — ML prediction,
ZBG pharmacophore matching, ADMET profiling, structural
novelty, and molecular docking — identified a final set
of priority compounds for experimental validation.
The top priority compounds include:

(i) STK989244 — best overall drug-like profile with high
GI absorption, QED of 0.646, and strong composite score;

(ii) STK651137 — strongest MM-GBSA binding energy
(-52.5 kcal/mol) with favorable drug-like properties;

(iii) PLANTAGINOSIDE A — strongest Glide XP docking score
(-11.607) representing a novel natural glycoside scaffold;

(iv) CNP0596213 — top COCONUT hydroxamic acid compound
with triple ZBG (hydroxamic acid, sulfonamide, N-hydroxy)
and predicted pIC50 of 8.547;

(v) Isochlorogenic acid A — novel natural polyphenol
scaffold with dual carboxylic acid-catechol ZBG, strong
docking score, and known safety profile in humans.

These compounds represent structurally diverse, mechanistically
credible LpxC inhibitor candidates spanning synthetic,
semi-synthetic, and natural product chemical spaces,
providing a strong basis for experimental validation through
biochemical LpxC enzyme inhibition assays and minimum
inhibitory concentration (MIC) determination against
P. aeruginosa.
"""

with open(project_path + r'\paper\results.txt',
          'w', encoding='utf-8') as f:
    f.write(results)

print("Results section saved ✅")
print(f"Word count: {len(results.split())} words")
```
